## Brief tutorial to handle multi-orbital Tight-binding models

```python
import dga.brillouin_zone as bz
import dga.wannier as wannier
```

```python
# load the wannier file in wannier90 format:

fname = 'path to your file'
hr = wannier.create_wannier_hr_from_file(fname)

# define the k-mesh:
nk = (32,32,1) # start small for multi-orbital models
sym = bz.two_dimensional_square_symmetries() # define the symmetries of your lattice
k_grid = bz.KGrid(nk=nk,symmetries=sym)
```

```python
# compute the multi-orbital tight-binding Hamiltonian in k-space
ek = hr.get_ek(k_grid,one_band=False)
```

```python
#save the Hamiltonian to file. Format is that used by w2dynamics as input:
fname_out = ' path and name of your output file'
ek_list = k_grid.map_fbz_mesh2list(ek)
wannier.write_hk_wannier90(ek_list,fname_out,k_grid.kmesh_list,nk)
```
